```python
import json
import matplotlib.pyplot as plt
```

```python
with open('./loss_info.json', 'r') as json_file :
    loss_info = json.load(json_file)

train_losses = [loss['train loss'] for loss in loss_info]
val_losses = [loss['val loss'] for loss in loss_info]
```

```python
plt.plot(train_losses)
plt.plot(val_losses)
```
